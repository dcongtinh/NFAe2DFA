```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.drawing.nx_agraph import to_agraph

# define the graph as per your question
# G = nx.MultiDiGraph([
#     ("A", "B", {'label': 'a'}), 
#     ("A", "C", {'label': 'b'}),
#     ("B", "B", {'label': 'a'}), 
#     ("B", "D", {'label': 'b'}),
#     ("C", "B", {'label': 'a'}), 
#     ("C", "C", {'label': 'b'}),
#     ("D", "B", {'label': 'a'}), 
#     ("D", "E", {'label': 'b'}),
#     ("E", "B", {'label': 'a'}), 
#     ("E", "C", {'label': 'b'}),
# ])

G = nx.MultiDiGraph()
# G.add_edge(' ', 'A', label='start', penwidth=2.0)
G.add_nodes_from(['A'], penwidth=2.0, color="blue")
G.add_nodes_from(['E'], penwidth=2.0, color="red")
G.add_edge(' ', 'A', label=' Start', penwidth=2.0)
G.add_edge('A', 'B', label=' a')
G.add_edge('A', 'C', label=' b')
G.add_edge('B', 'B', label=' a')
G.add_edge('B', 'D', label=' b')
G.add_edge('C', 'B', label=' a')
G.add_edge('C', 'C', label=' b')
G.add_edge('D', 'B', label=' a')
G.add_edge('D', 'E', label=' b')
G.add_edge('E', 'B', label=' a')
G.add_edge('E', 'C', label=' b')


# add graphviz layout options (see https://stackoverflow.com/a/39662097)
G.graph['edge'] = {'arrowsize': '0.6', 'splines': 'curved'}
G.graph['graph'] = {'scale': '14'}

# adding attributes to edges in multigraphs is more complicated but see
# https://stackoverflow.com/a/26694158
# G["B"]["B"]['color'] = 'red'

A = to_agraph(G)
A.layout('dot')
A.draw('multi.png')

import cv2
img = plt.imread('multi.png')
plt.axis("off")
plt.imshow(img)
```
