## Import class NFAe

```python
from FA import NFAe
```

## Nhập Input

```python
states = {1, 2, 3}  # trạng thái states
alphabet = ['a', 'b', 'ε']  # bộ chữ cái nhập alphabet

# Hàm chuyển trạng thái
transition_function = {
    (1, 'a'): {3},
    (1, 'ε'): {2},
    (2, 'a'): {1},
    (3, 'a'): {2},
    (3, 'b'): {2, 3},
}
start_state = {1}  # trạng thái bắt đầu
accept_states = {2}  # trạng thái kết thúc
```

## Khởi tạo lớp NFAe

```python
nfae = NFAe(states, alphabet, transition_function,
            start_state, accept_states)
```

## Chuyển NFAe sang DFA

```python
nfae.toDFA()
```

## Bảng hàm chuyển của NFAe

```python
nfae.printNFAeFuncTable()
```

## Sơ đồ chuyển của NFAe

```python
nfae.printGraph('NFAe')
```

## Bảng hàm chuyển của DFA

```python
nfae.printDFAFuncTable()
```

## Sơ đồ chuyển của DFA

```python
nfae.printGraph('DFA')
```
